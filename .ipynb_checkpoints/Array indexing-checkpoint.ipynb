#### Numpy indexing and Slicing of vectors

```python
arr = np.arange(11,21)
arr
```

indexing is smilar to python list starts from 0 and end at numberOfElement-1

```python
print(arr[6])
```

##### Slicing

let we want to slice from 12 to 15 or index 1 to 4  
syntax : arr[start:end+1:steps]  
by default start  =0  
end = end  
steps = 1

```python
arr[1:5]
```

if we write arr[:5] then default value set to 0, and if we do not provide the end one then it also go till the end

```python
arr[1:7:2]
```

#### Numpy indexing and slicing of matrix

```python
arr2 = np.arange(1,31).reshape(6,5)
arr2
```

arr2[n]  it gives the n'th array(row) 

```python
arr2[4]
```

arr2[n,m] gives m'th element of n'th row

```python
arr2[1,2]
```

```python
slice = arr2[0:2,1:3]
slice
```

**Here we give arr[slicing of rows, slicing of cols]**

```python
arr2[3:,3:]
```

if you want all rows, but sliced cols then

```python
arr2[:,2]
```

it gives the third coloumn
