#### Vectors and Metrices

Vectors are one dimentional, where Metrices are multi-dimentional   
We can store vector in list but for metrices we need row and coloumns which are not supported by lists so we make, multidimensional arrays

```python
l1 = [[1,2,3],[4,5,8],[10,43,56]]
l1
```

**Above you can see it, it is looking one dimentional**  
Now we will convert it to array

```python
np.array(l1)
```

*So here we can see that its looking like a metrix*

#### Array Attributes 

```python
arr11 = np.array([[7,2,3],[9,56,20],[43,54,56]])
arr11
```

```python
arr11.shape    #it tells the shape of the array(row, column)
```

```python
arr11.size   #tells the number of elements in the array
```

```python
arr11.dtype      #tells the dataType of the array
```

#### Array Methods

```python
# we already have a array 
arr11
```

```python
arr11.min()  #tells the minimum element of the array 
```

```python
arr11.max()   #tells the maximum element of the array 
```

```python
arr11.sum()   #sum all the elemnts of the array
```

If you want sum of rows and columns then --->

```python
np.sum(arr11, axis = 0)
```

in axis =0  it sums each coloumn separetly

```python
np.sum(arr11, axis = 1)
```

in axis = 1 it sums each row separetly

```python
arr11.mean()  #gives the middle elemnet(sum of elemnts)/total num of element
```

```python
arr11.std()   #tells the standrd deviation
```

```python
arr11.argmax()    #tells the index of maximum element
```

```python
arr11.argmin()   #tells the index of minimum element
```

#### Resizing and Reshaping

```python
arr12 = np.arange(1,31)
arr12
```

**Its a 1D array and we want to convert it into vector (2D array)**  
If you are reshaping something then your element count should be the same

```python
arr12 = arr12.reshape(6,5)
arr12
```

so all our 30 elements are utilized
