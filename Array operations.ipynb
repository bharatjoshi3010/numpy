#### Broadcasting 

If you have a single number and you want to add (perform any opertion) this number to each element of the array then

```python
l1 = [2,4,5,6,9,10]
arr3 = np.array(l1)
arr3
```

adding 10 in each element of arr3

```python
arr3 + 10                              #it is called broadcasting
```

**Performing these in vector**

```python
arr4 = np.arange(1,26).reshape(5,5)
arr4
```

```python
arr5 = np.arange(3,28).reshape(5,5)
arr5
```

```python
arr4+arr5
```

Broadcasting in vactor

```python
arr5+10
```

#### Deep And Shallow Copy

```python
arr6 = np.arange(1,21)
arr6
```

```python
slice = arr6[:5]
slice = slice * 10                   #performed broadcasting in slice and also saved it
slice
```

```python
arr6
```

So here we can see that the changes in slice are not affecting the actual array, but if we change slice of a list then it will also affect the actual list also , That means array makes **Shallow copy** when we slice them and occupy new space for the slice

```python
a = np.arange(1,10)
a
```

```python
b=a
```

```python
b[0] = 99
```

```python
a
```

here we can see that the change in **b** also changing the **a**,  
so by b=a they just pass on the locations, **b** will not occupy a new place for him(its a **deep copy**)  
if you want to make a shallow copy then write b = a.copy()

```python
c = np.arange(10,20);
c
```

```python
d=c.copy()
```

```python
d[1] = 77
```

```python
d
```

```python
c
```

here you can see there are no changes in **c**
