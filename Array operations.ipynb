#### Broadcasting 

If you have a single number and you want to add (perform any opertion) this number to each element of the array then

```python
l1 = [2,4,5,6,9,10]
arr3 = np.array(l1)
arr3
```

adding 10 in each element of arr3

```python
arr3 + 10                              #it is called broadcasting
```

**Performing these in vector**

```python
arr4 = np.arange(1,26).reshape(5,5)
arr4
```

```python
arr5 = np.arange(3,28).reshape(5,5)
arr5
```

```python
arr4+arr5
```

Broadcasting in vactor

```python
arr5+10
```

#### Deep And Shallow Copy

```python
arr6 = np.arange(1,21)
arr6
```

```python
slice = arr6[:5]
slice = slice * 10                   #performed broadcasting in slice and also saved it
slice
```

```python
arr6
```

So here we can see that the changes in slice are not affecting the actual array, but if we change slice of a list then it will also affect the actual list also , That means array makes **Shallow copy** when we slice them and occupy new space for the slice

```python
a = np.arange(1,10)
a
```

```python
b=a
```

```python
b[0] = 99
```

```python
a
```

here we can see that the change in **b** also changing the **a**,  
so by b=a they just pass on the locations, **b** will not occupy a new place for him(its a **deep copy**)  
if you want to make a shallow copy then write b = a.copy()

```python
c = np.arange(10,20);
c
```

```python
d=c.copy()
```

```python
d[1] = 77
```

```python
d
```

```python
c
```

here you can see there are no changes in **c**

#### Advance Array Manipulation

**Stacking Array**

We can stack two arrays(1D, vectors only) only not to matrics 

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b))            #it will make vertical stack
```

```python
np.hstack((a,b))             #it will make horizental stack (two array get joined)
```

```python
np.column_stack((a,b))            #it will make a coloumn stack 
```

**Splitting Array**

```python
c = np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c,2)         #np.hsplit(array, n), it splits array horizentaly in 'n' parts
```

here the number of column should be divisible by 'n'

```python
np.vsplit(c, 2)                  #np.vsplit(array, n),it will do vertical split
```

here the number of row should be divisible by 'n'
