# Array Operations

```python
import numpy as np
```

#### Arithmetic Operation

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
```

**Addition of array having same number of elements**

```python
arr2 = a1 + a2                   #size of both array must be same
arr2
```

All the elements having same index added together

**Substraction**  
array having same number of elements

```python
a1 - a2                      #size of both array must be same
```

similary we can peerform *, /, //, **

#### Broadcasting 

If you have a single number and you want to add (perform any opertion) this number to each element of the array then

```python
l1 = [2,4,5,6,9,10]
arr3 = np.array(l1)
arr3
```

adding 10 in each element of arr3

```python
arr3 + 10                              #it is called broadcasting
```

**Performing these in vector**

```python
arr4 = np.arange(1,26).reshape(5,5)
arr4
```

```python
arr5 = np.arange(3,28).reshape(5,5)
arr5
```

```python
arr4+arr5
```

Broadcasting in vactor

```python
arr5+10
```

#### Deep And Shallow Copy

```python
arr6 = np.arange(1,21)
arr6
```

```python
slice = arr6[:5]
slice = slice * 10                   #performed broadcasting in slice and also saved it
slice
```
