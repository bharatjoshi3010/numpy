#### Vectors and Metrices

Vectors are one dimentional, where Metrices are multi-dimentional   
We can store vector in list but for metrices we need row and coloumns which are not supported by lists so we make, multidimensional arrays

```python
l1 = [[1,2,3],[4,5,8],[10,43,56]]
l1
```

**Above you can see it, it is looking one dimentional**  
Now we will convert it to array

```python
np.array(l1)
```

*So here we can see that its looking like a metrix*
